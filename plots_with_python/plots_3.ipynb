```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from feynman import Diagram
from feynman import Line

class Theme:
    theme = "light"
    @staticmethod
    def set(_theme):
        Theme.theme = _theme
        if(Theme.theme == "dark"):
            plt.style.use('dark_background')
        else:
            plt.style.use('default')
        plt.rcParams['font.family'] = 'serif'
        plt.rcParams['font.serif'] = ['cmr10']  # Основной шрифт для текста
        plt.rcParams['mathtext.fontset'] = 'cm'
    @staticmethod
    def color():
        if(Theme.theme == "light"):
            return "black"
        else:
            return "white"
    @staticmethod
    def back():
        if(Theme.theme == "light"):
            return "white"
        else:
            return "black"
    @staticmethod
    def suff():
        if(Theme.theme == "light"):
            return "l"
        else:
            return "d"
```

```python
Theme.set("light")
#Theme.set("dark")
main_color = Theme.color()
style_params = {"color": main_color}
arrow_params = {"color": main_color,"width":0.02,"length":0.05}
fig = plt.figure(figsize=(3*2.4,2.4))
ax = fig.add_axes([0,0,1,1 ], frameon=False)
diagram = Diagram(ax)

ax.set_xlim(0, 3*0.4)
ax.set_ylim(0, 0.4)

H = 0.4
W = 0.4

###############
V_x = diagram.vertex(xy=(W/2-0.02, H-0.04), **style_params,markersize=6)
V_y = diagram.vertex(xy=(W/2+0.02, H-0.04), **style_params,markersize=6)

V_x1 = diagram.vertex(xy=(0.05, 0.04), marker='', **style_params)
V_x2 = diagram.vertex(xy=(W-0.05, 0.04), marker='', **style_params)

diagram.line(V_x,V_x1, **style_params,linewidth=2, arrow_param=arrow_params,arrow=False)
diagram.line(V_y,V_x2, **style_params,linewidth=2,arrow_param=arrow_params,arrow=False)


diagram.text(V_x.xy[0]-0.04, V_x.xy[1]+0.01, r"$x$", fontsize=32)
diagram.text(V_y.xy[0]+0.04, V_y.xy[1]+0.01, r"$y$", fontsize=32)


diagram.text(V_x1.xy[0]+0.05, V_x1.xy[1]+0.01, r"$x_1$", fontsize=32)
diagram.text(V_x2.xy[0]-0.05, V_x2.xy[1]+0.01, r"$x_2$", fontsize=32)

############

U_x = diagram.vertex(xy=(W+W/2-0.02, H-0.04), **style_params,markersize=6)
U_y = diagram.vertex(xy=(W+W/2+0.02, H-0.04), **style_params,markersize=6)

U_x1 = diagram.vertex(xy=(W+0.05, 0.04), marker='', **style_params)
U_x2 = diagram.vertex(xy=(W+W-0.05, 0.04), marker='', **style_params)

diagram.line(U_x,U_x2, **style_params,linewidth=2, arrow_param=arrow_params,arrow=False)
diagram.line(U_y,U_x1, **style_params,linewidth=2,arrow_param=arrow_params,arrow=False)


diagram.text(U_x.xy[0]-0.04, U_x.xy[1]+0.01, r"$x$", fontsize=32)
diagram.text(U_y.xy[0]+0.04, U_y.xy[1]+0.01, r"$y$", fontsize=32)

diagram.text(U_x1.xy[0]+0.05, U_x1.xy[1]+0.01, r"$x_1$", fontsize=32)
diagram.text(U_x2.xy[0]-0.05, U_x2.xy[1]+0.01, r"$x_2$", fontsize=32)

###########

# add arrow to ax

diagram.text(2*W, H/2, r"$\Longrightarrow$", fontsize=32)

###########

W_y = diagram.vertex(xy=(2*W+W/2, H-0.04), **style_params,markersize=12)

W_x1 = diagram.vertex(xy=(2*W+0.05, 0.04), marker='', **style_params)
W_x2 = diagram.vertex(xy=(2*W+W-0.05, 0.04), marker='', **style_params)

diagram.line(W_y,W_x2, **style_params,linewidth=2, arrow_param=arrow_params,arrow=False)
diagram.line(W_y,W_x1, **style_params,linewidth=2,arrow_param=arrow_params,arrow=False)

diagram.text(W_y.xy[0]+0.04, W_y.xy[1]+0.01, r"$y$", fontsize=32)

diagram.text(W_x1.xy[0]+0.05, W_x1.xy[1]+0.01, r"$x_1$", fontsize=32)
diagram.text(W_x2.xy[0]-0.05, W_x2.xy[1]+0.01, r"$x_2$", fontsize=32)

diagram.plot()


fig.savefig(f'../docs/DeepInelastic/OPE/images/OPE_phiphi-{Theme.suff()}.svg',transparent=True)
```

```python
Theme.set("light")
#Theme.set("dark")
main_color = Theme.color()
style_params = {"color": main_color}
arrow_params = {"color": main_color,"width":0.02,"length":0.05}
fig = plt.figure(figsize=(3*2.4,2.4))
ax = fig.add_axes([0,0,1,1 ], frameon=False)
diagram = Diagram(ax)

ax.set_xlim(0, 3*0.4)
ax.set_ylim(0, 0.4)


H = 0.4

W_shift = 0.15

diagram.text( 0.09, H/2, r"$\int {\; \operatorname{d}\! \Pi_X}$", fontsize=24)

in1_L = diagram.vertex(xy=(W_shift+ 0.05, H-0.08), marker='', **style_params)
in2_L = diagram.vertex(xy=(W_shift +0.05, 0.08), marker='', **style_params)


diagram.text( in1_L.xy[0]+0.05, in1_L.xy[1]+0.01, r"$e^{\!\!\!-}$", fontsize=32)
diagram.text( in2_L.xy[0]+0.05, in2_L.xy[1]-0.01, r"$e^{\!\!\!+}$", fontsize=32)

Module_ld = diagram.vertex(xy=(in2_L.xy[0]-0.02, 0.03), marker='', **style_params)
Module_lu = diagram.vertex(xy=(in2_L.xy[0]-0.02, H-0.03), marker='', **style_params)
diagram.line(Module_ld,Module_lu, **style_params,linewidth=1.5, arrow_param=arrow_params,arrow=False)

Module_rd = diagram.vertex(xy=(W_shift+0.34+0.02, 0.03), marker='', **style_params)
Module_ru = diagram.vertex(xy=(W_shift+0.34+0.02, H-0.03), marker='', **style_params)
diagram.line(Module_rd,Module_ru, **style_params,linewidth=1.5, arrow_param=arrow_params,arrow=False)
diagram.text( Module_ru.xy[0]+0.03, Module_ru.xy[1]-0.01, r"$2$", fontsize=26)


center1_L = diagram.vertex(xy=(W_shift+0.12, 0.2), **style_params,markersize=4)
center2_L = diagram.vertex(xy=(W_shift+0.26, 0.2),**style_params,markersize=4)

diagram.line(in1_L,center1_L, **style_params,linewidth=2, arrow_param=arrow_params)
diagram.line(center1_L, in2_L,**style_params,linewidth=2, arrow_param=arrow_params)
diagram.line(center1_L,center2_L, **style_params,linewidth=2, style='wiggly',nwiggles = 3)

vs = [diagram.vertex(xy=(W_shift+0.34,H/2 + 0.03*i), marker='', **style_params) for i in range(-3,4)]
for v in vs:
    diagram.line(center2_L,v, **style_params,linewidth=1.5, arrow_param=arrow_params,arrow=False) 

diagram.text( W_shift+0.3, H-0.1, r"$X$", fontsize=26)


diagram.text( W_shift+0.46, H/2, r"$=2\operatorname{Im}$", fontsize=26)



W_center = 0.94
W = 0.24


in_1 = diagram.vertex(xy=(W_center-W, H-0.08), marker='', **style_params)
in_2 = diagram.vertex(xy=(W_center-W, 0.08), marker='', **style_params)



diagram.text( in_1.xy[0]+0.05, in_1.xy[1]+0.01, r"$e^{\!\!\!-}$", fontsize=32)
diagram.text( in_2.xy[0]+0.05, in_2.xy[1]-0.01, r"$e^{\!\!\!+}$", fontsize=32)

center_1 = diagram.vertex(xy=(W_center-2*W/3, 0.2), **style_params,markersize=4)
center_2 = diagram.vertex(xy=(W_center+2*W/3, 0.2),**style_params,markersize=4)

center_3 = diagram.vertex(xy=(W_center-W/6, 0.2),**style_params,markersize=4)
center_4 = diagram.vertex(xy=(W_center+W/6, 0.2),**style_params,markersize=4)

out_1 = diagram.vertex(xy=(W_center+W, H-0.08), marker='', **style_params)
out_2 = diagram.vertex(xy=(W_center+W, 0.08), marker='', **style_params)



diagram.text( out_1.xy[0]-0.05, out_1.xy[1]+0.01, r"$e^{\!\!\!-}$", fontsize=32)
diagram.text( out_2.xy[0]-0.05, out_2.xy[1]-0.01, r"$e^{\!\!\!+}$", fontsize=32)

diagram.line(in_1,center_1, **style_params,linewidth=2, arrow_param=arrow_params)
diagram.line(center_1, in_2,**style_params,linewidth=2, arrow_param=arrow_params)

diagram.line(center_1,center_3, **style_params,linewidth=2, style='wiggly',nwiggles = 3)
diagram.line(center_2,center_4, **style_params,linewidth=2, style='wiggly',nwiggles = 3)
ax.add_patch(
    mpl.patches.Circle(
        (W_center, H / 2),
        0.04,
        edgecolor=main_color,
        facecolor=mpl.colors.to_rgba(main_color, alpha=0.5),
        linewidth=1.5,
    )
)

diagram.line(center_2,out_1, **style_params,linewidth=2, arrow_param=arrow_params)
diagram.line(center_2,out_2, **style_params,linewidth=2, arrow_param=arrow_params)

diagram.plot()

#fig.savefig(f'../docs/DeepInelastic/OPE/images/OPE_OpticalTheorem-{Theme.suff()}.svg',transparent=True)
```
